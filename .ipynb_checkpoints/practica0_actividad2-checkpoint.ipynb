## 1. Identificadores de variables

Los identificadores de variables en Python comienzan con una letra o underscore y están seguidas de cualquier letra, número y underscores.

```python
a = "variable tipo string"
valorInicial = 1  # variable tipo entero y un comentario

print(a)
print(valorInicial)
```

## 2. Tipos numéricos

Los tipos numéricos en Python incluyen: enteros, punto flotante y números complejos.

```python
a = 1        # entero
b = 3.4      # punto flotante
c = 2+3j     # número complejo

print(a, b, c)
print(f"Tipo de a: {type(a)}")
print(f"Tipo de b: {type(b)}")
print(f"Tipo de c: {type(c)}")
```

## 3. Cadenas de texto

Las constantes de cadenas pueden escribirse entre `" "` y `' '`. Podemos escribir cadenas que abarquen más de un renglón entre `"""` y `"""`.

```python
s1 = "Esto es una cadena con retorno de carro\n"
s2 = 'También podemos usar las "comillas sencillas"\n'
s3 = """ Esta cadena
    abarca varios
    renglones ... """

print(s1, s2, s3)
```

## 4. Listas

Las listas en Python se escriben entre corchetes con sus elementos separados por comas.

```python
a = [1, 3, 25]
b = [1] * 10
c = list()
d = []

print("a =", a)
print("b =", b)
print("c =", c)
print("d =", d)
```

## 5. Función range()

Podemos usar la función `range()` para crear listas. Recuerda que el extremo derecho no es inclusivo.

```python
a = list(range(4))           # 0, 1, 2, 3
b = list(range(5, 10))       # 5, 6, 7, 8, 9
c = list(range(0, 100, 10))  # 0, 10, 20, ..., 90
d = list(range(10, 0, -1))   # 10, 9, 8, ..., 1

print("a =", a)
print("b =", b)
print("c =", c)
print("d =", d)
```

## 6. Indexación de listas

Usamos corchetes para indexar elementos y porciones de una lista. El extremo derecho tampoco es inclusivo cuando usamos `:`.

```python
a = list(range(1, 100))

print("Elementos individuales:")
print(f"a[0]={a[0]}, a[1]={a[1]}, a[10]={a[10]}, a[25]={a[25]}, a[50]={a[50]}")

print("\nPrimeros 10 elementos:")
print(a[:10])

print("\nÚltimos elementos (desde índice 90):")
print(a[90:])

print("\nElementos del índice 25 al 49:")
print(a[25:50])
```

## 7. Sentencias if

Podemos tomar decisiones con las sentencias `if`, `elif` y `else`. Es importante la indentación para definir los bloques de código.

```python
a = 5

if a < 10:
    print("menor a 10")
elif a > 10:
    print("mayor a 10")
else:
    print("igual a 10")
```

```python
# Probemos con otros valores
a = 10

if a < 10:
    print("menor a 10")
elif a > 10:
    print("mayor a 10")
else:
    print("igual a 10")
```

## 8. Ciclos for y while

Python cuenta con ciclos `for` y `while` para iteraciones.

```python
# Iterar sobre una lista
lista = [1, 2, 3]
print("Iterando sobre lista:")
for i in lista:
    print(i)
```

```python
# Iterar con range ascendente
print("Conteo de 0 a 10:")
for i in range(0, 11):
    print(i, end=" ")
```

```python
# Iterar con range descendente
print("Conteo regresivo de 10 a 0:")
for i in range(10, -1, -1):
    print(i, end=" ")
```

```python
# Ciclo while
print("Usando while:")
contador = 0
while contador <= 10:
    print(contador, end=" ")
    contador += 1
```

## 9. Funciones

Podemos definir funciones con la palabra clave `def`.

```python
def cuadrado(n):
    return n * n

def factorial(n):
    if n <= 1:
        return 1
    else:
        return n * factorial(n - 1)

print(f"cuadrado(5) = {cuadrado(5)}")
print(f"factorial(5) = {factorial(5)}")
```
