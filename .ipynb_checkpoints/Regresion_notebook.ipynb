```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
from sklearn import linear_model 
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv("/Users/enriquemascarua/Downloads/Copier Sales of America.csv")
```

```python
data.shape
```

```python
data.head
```

```python
x_train = np.array(data['llamadas']).reshape(-1,1)
y_train = np.array(data['ventas']) 
```

```python
r = linear_model.LinearRegression()
r.fit(x_train, y_train)
y_pred = r.predict(x_train)
print(f"correlacion: {r.coef_}")
print(f"Interseccion: {r.intercept_}")
print(f"Error cuadratico: {mean_squared_error(y_train, y_pred)}")
print(f"Varianza: {r2_score(y_train, y_pred)}")
```
